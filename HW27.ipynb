```python
import numpy as np
import pandas as pd
mobile_df = pd.read_csv(r"mobile.csv")
mobile_df
```

```python
def find_first_name(full_name):
    i = 0
    first_name = ""
    while i <= (len(full_name) - 1):
      if not full_name[i] == " ":
         first_name += full_name[i]
         i += 1
      else:
         break
    return first_name
mobile_df['First Name'] = mobile_df['Customer Name'].apply(find_first_name)
mobile_df
```

```python
mobile_df['Phone pre'] = mobile_df['Phone Number'].apply(lambda value: str(value)[0:3])
mobile_df
mobile_df['Phone pre'] = mobile_df['Phone pre'].replace(['050', '051', '052', '053', '054'],['Pelephone', 'Golan', 'Cellcom', 'Motorola', 'Hot Mobile'])
mobile_df
```
